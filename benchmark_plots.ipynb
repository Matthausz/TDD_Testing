```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Runtime Scalling For Qubits, Gate count, and CNOT gate count

```python

# Read the benchmark results file into a DataFrame
df = pd.read_csv("statevector_benchmark.txt")

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Number of Qubits'], df['Time (s)'], c='blue', alpha=0.5)
plt.title('Scatter Plot of Time vs Number of Qubits for Statevector Method')
plt.xlabel('Number of Qubits')
plt.ylabel('Time (s)')
plt.grid(True)
plt.show()

# Plot the scatter plot
plt.figure(figsize=(10, 6))
scatter=plt.scatter(df['Number of Gates'], df['Time (s)'], c=df['Number of Qubits'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Number of Qubits')
plt.title('Scatter Plot of Time vs Number of Gates for Statevector Method')
plt.xlabel('Number of Gates')
plt.ylabel('Time (s)')
plt.yscale("log")
plt.xscale("log")
plt.grid(True)
plt.show()

# Plot the scatter plot
plt.figure(figsize=(10, 6))
scatter=plt.scatter(df['Number of CNOT Gates'], df['Time (s)'], c=df['Number of Qubits'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Number of Qubits')
plt.title('Scatter Plot of Time vs Number of CNOT Gates for Statevector Method')
plt.xlabel('Number of CNOT  Gates')
plt.ylabel('Time (s)')
plt.yscale("log")
plt.xscale("log")
plt.grid(True)
plt.show()
```

```python
#QFT only runs
df = pd.read_csv("statevector_benchmark.txt")
# Filter out examples that don't have a circuit name starting with "QFT"
df = df[df['Circuit Name'].str.startswith('qft')]

# Convert "TO" values to NaN for easier handling
df['Time (s)'] = pd.to_numeric(df['Time (s)'], errors='coerce')

# Generate scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Number of Gates'], df['Time (s)'], c=df['Number of Qubits'], cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Number of Qubits')
plt.xlabel('Number of Gates')
plt.ylabel('Time (s)')
plt.title('Time vs Number of Gates For QFT with Statevector')
plt.grid(True)
plt.show()
```

# Below I try to find what circuits cause the upper branch in the runtime plots

This showed that for some reason quantum_volume and qft circuit architectures scalled worse than the other algorithims.

```python
#only slow runs
df = pd.read_csv("TDD_benchmark.txt")
# Filter out examples that don't have a circuit name starting with "QFT"
df = df[df['Time (s)']>6]

# Convert "TO" values to NaN for easier handling
df['Time (s)'] = pd.to_numeric(df['Time (s)'], errors='coerce')

# Generate scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Number of Gates'], df['Time (s)'], c="red", alpha=0.7)
plt.xlabel('Number of Gates')
plt.ylabel('Time (s)')
plt.title('Time vs Number of Gates For QFT with TDD')
# Label each dot with the circuit name
for i, txt in enumerate(df['Circuit Name']):
    plt.text(df['Number of CNOT Gates'].iloc[i], df['Time (s)'].iloc[i], txt, ha='center', va='center', fontsize=8)
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("TDD_benchmark.txt")
# Convert "TO" values to NaN for easier handling
df['Time (s)'] = pd.to_numeric(df['Time (s)'], errors='coerce')

# Filter out examples that arent correct circuit type
df1 = df[df['Circuit Name'].str.startswith('quantum_volume')]
df2 = df[df['Circuit Name'].str.startswith('qft')]

# Generate scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df1['Number of Gates'], df1['Time (s)'], c="red", alpha=0.7)
plt.scatter(df2['Number of Gates'], df2['Time (s)'], c="yellow", alpha=0.7)
scatter=plt.scatter(df['Number of Gates'], df['Time (s)'], c=df['Number of Qubits'], cmap='viridis', alpha=0.27)
plt.colorbar(scatter, label='Number of Qubits')
plt.xlabel('Number of Gates')
plt.ylabel('Time (s)')
plt.legend(["Quantum_Volume","QFT","Other"])
plt.title('Time vs Number of Gates For QFT with TDD')
plt.yscale("log")
plt.xscale("log")
plt.grid(True)
plt.show()
```
